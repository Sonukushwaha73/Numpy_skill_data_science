# linear algebra

# formula

```python
|a b|
|c d| = ad − bc


    = (1×4) − (2×3)
= 4 − 6
= -2



Index = d*(R*C) + r*C + c
d = block number (0 se start)

R = rows per block

C = columns per row

r = row number

c = column number

Matlab: block → row → column ke hisaab se flat memory index nikalta hai


arr = np.array([
    [[1, 2, 3],       # Block 0, Row 0
     [4, 5, 6]],      # Block 0, Row 1

    [[7, 8, 9],       # Block 1, Row 0
     [10, 11, 12]]    # Block 1, Row 1
])
Shape = (2, 2, 3) → D=2, R=2, C=3
d*(R*C) + r*C 
arr[0,0,0] = 0*(2*3) + 0*3 + 0 = 0  → value 1
arr[0,0,1] = 0*(2*3) + 0*3 + 1 = 1  → value 2
arr[0,0,2] = 0*(2*3) + 0*3 + 2 = 2  → value 3

arr[0,1,0] = 0*(2*3) + 1*3 + 0 = 3  → value 4
arr[0,1,1] = 0*(2*3) + 1*3 + 1 = 4  → value 5
arr[0,1,2] = 0*(2*3) + 1*3 + 2 = 5  → value 6
```
