# swapaxes

```python
#  row --columns
# column----row


# swapaxes() kya karta hai?

#  swapaxes() ka kaam hota hai array ke do axes (dimensions) ko ek dusre ke saath swap karna.

# Matlab:

# 2D me → rows aur columns ko swap karna (transpose ka kaam karta hai)

# 3D+ me → do axes ko interchange karna

# 🔹 Syntax:
# np.swapaxes(array, axis1, axis2)
```

```python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

swapped = np.swapaxes(arr, 0, 1)#swapaxes(arr, 0, 1) → rows aur columns swap ho gaye
print(swapped)
```

# inverse

```python
# np.linalg.inv() kya karta hai?

#  inverse() ka kaam hota hai matrix ka ulta (inverse) nikalna.

# Sirf square matrix (rows = columns) ka inverse hota hai.

# Agar matrix ka determinant 0 ho → inverse exist nahi karta.

# 🔹 Syntax:
# np.linalg.inv(matrix)

# Part	Matlab
# matrix	Jo square matrix hai jiska inverse chahiye
```

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

inv_A = np.linalg.inv(A)
print(inv_A)
```

# power

# Determinate

```python
# Determinant kya hai?

# Determinant ek square matrix ke liye ek number hai jo matrix ke properties ke baare me batata hai:

# Agar determinant = 0 → matrix invertible nahi hai

# Agar determinant ≠ 0 → matrix invertible hai

# 🔹 NumPy me Syntax:
# np.linalg.det(matrix)

# Part	Matlab
# matrix	Jo square matrix ka determinant chahiye
```

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

det_A = np.linalg.det(A)
print(det_A)
```
